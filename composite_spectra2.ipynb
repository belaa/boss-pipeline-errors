```python
def get_rest_frame(wlen_list,flux_list,ivar_list,z_list):
    rest_wlen = []
    rest_flux = []
    rest_ivar = []
    for i in range(len(wlen_list)):
        a_temp = 1/(1+z_list[i])
        rest_wlen.append(multiply(wlen_list[i],a_temp))
        rest_flux.append(divide(flux_list[i],a_temp))
        rest_ivar.append(divide(ivar_list[i],a_temp))
    return rest_wlen,rest_flux,rest_ivar
```

```python
def get_flux1450(wlen_rest_ordered_list,flux_rest_ordered_list):
    flux_1450 = ['x']*len(wlen_rest_ordered_list)
    index_1450 = ['x']*len(wlen_rest_ordered_list)
    for i in range(len(wlen_rest_ordered_list)):
        for j in range(len(wlen_rest_ordered_list[i])):
            if 1448.0<wlen_rest_ordered_list[i][j]<1452.0:
                flux_1450[i] = flux_rest_ordered_list[i][j]
                index_1450[i] = i
                break
    return flux_1450,index_1450
```

```python
def normalize_spectra(flux_1450,flux_rest_ordered_list):
    mean_flux = 1.0 #mean(flux_1450)
    x = 0 #value at 1450 angstroms
    beta = 0
    new_flux = []
    for i in range(len(flux_rest_ordered_list)):
        if flux_1450[i]!='x':
            x = flux_1450[i]
            beta = mean_flux/x
            new_flux.append(multiply(beta,flux_rest_ordered_list[i]))
        else:
            new_flux.append(flux_rest_ordered_list[i])
    return new_flux
```

```python
def find_wlen_range(wlen_rest):
    min_wlen = []
    max_wlen = []
    for i in range(len(wlen_rest)):
        min_wlen.append(min(wlen_rest[i]))
        max_wlen.append(max(wlen_rest[i]))
    return min(min_wlen),max(max_wlen)
```

```python
def get_all_exposures(data_file):
    meta_data = Table.read(data_file,format='ascii')
    finder = path.Finder()
    mirror = remote.Manager()
    PLATE,MJD,FIBER = meta_data['PLATE'],meta_data['MJD'],meta_data['FIBER']
    WLEN = []
    FLUX = []
    DFLUX = []
    Z = meta_data['Z_VI']
    new_z = []
    x = []
    for i in range(10000,10020):
        x.append(i)
    for j in range(40000,40070):
        x.append(j)
    for k in range(60000,60070):
        x.append(k)
    for l in range(70000,70020):
        x.append(l)
    for i in range(len(x)):
        remote_path = finder.get_spec_path(PLATE[x[i]],MJD[x[i]],FIBER[x[i]],lite=False)
        local_path = mirror.get(remote_path)
        spec_obj = spec.SpecFile(local_path)
        data = spec_obj.get_valid_data()
        wlen,flux,dflux = data['wavelength'][:],data['flux'][:],data['dflux'][:]            
        WLEN.append(wlen)
        FLUX.append(flux)
        DFLUX.append(dflux)
        new_z.append(Z[x[i]])
    wlen_rest,flux_rest,ivar_rest = get_rest_frame(WLEN,FLUX,DFLUX,new_z)
    flux_1450,index_1450 = get_flux1450(wlen_rest,flux_rest)
    normalized_flux = normalize_spectra(flux_1450,flux_rest)
    #min_wlen,max_wlen = find_wlen_range(wlen_rest)
    min_wlen = 850
    max_wlen = 3000
    aligned_wlen,aligned_flux,aligned_ivar = align_wlen(wlen_rest,normalized_flux,ivar_rest,min_wlen,max_wlen)
    zero,first,second = bin_data(aligned_wlen,aligned_flux,aligned_ivar,min_wlen,max_wlen,1200)
    print min_wlen
    print max_wlen
    return zero,first,aligned_wlen,aligned_flux
```

```python
def smooth(x,beta):
 """ kaiser window smoothing """
 window_len=50
 # extending the data at beginning and at the end
 # to apply the window at the borders
 s = numpy.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
 w = numpy.kaiser(window_len,beta)
 y = numpy.convolve(w/w.sum(),s,mode='valid')
 return y[5:len(y)-5]
```
